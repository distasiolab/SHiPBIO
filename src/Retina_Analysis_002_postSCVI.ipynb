```python
filename = '/home/mdistasio/YaleGoogleDrive/DiStasio Lab Share/02 Analysis/annData_ManualAnnotate/calc/anndata_integrated.h5ad'
adata = ad.read_h5ad(filename)
```

```python




sq.pl.spatial_scatter(adata[np.where(adata.obs['Retina_1'])], color="leiden", size=1, shape=None)
sq.pl.spatial_scatter(adata[np.where(adata.obs['Retina_2'])], color="leiden", size=1, shape=None)
sq.pl.spatial_scatter(adata[np.where(adata.obs['Retina'])], color="leiden", size=1, shape=None)
```

```python
sc.pp.log1p(adata)

## Clustering
print('Clustering...')
sc.tl.pca(adata, svd_solver='arpack', n_comps=30)
sc.pp.neighbors(adata, n_neighbors=20, n_pcs=30)
sc.tl.leiden(adata)
sc.tl.paga(adata)
sc.pl.paga(adata, plot=False)
sc.tl.umap(adata, init_pos='paga')

print('Done!')
```
